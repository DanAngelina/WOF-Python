```python
import random
import time

letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
vowels  = 'AEIOU'
vowel_cost = 250
phrases = ["speak of the devil", "the best of both worlds", "see eye to eye", "when pigs fly", "costs an arm and a leg", "once in a blue moon", "a piece of cake", "let the cat out of the bag", "feeling under the weather", "kill two birds with one stone"]
wheel_segments = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 950, "Lose a Turn", "Bankrupt"]

class Player:
    def __init__(self, name):
        self.name = name
        self.prizeMoney = 0

    def addMoney( self, amt):
        self.prizeMoney += amt

    def goBankrupt(self):
        self.prizeMoney = 0

    def __str__(self):
        return "{} (${})".format(self.name,self.prizeMoney)

    def getGuess(self, hiddenPhrase, guessed):
        print("{} has (${})".format(self.name,self.prizeMoney))
        print("Phrase:",hiddenPhrase)
        print("Guessed:",guessed)

        Instructions = str(input("Guess a letter or entire phrase: "))
        return Instructions
def getphrase():
    randPhrase = random.choice(phrases)
    phrases.remove(randPhrase)
    return randPhrase.upper()

def hiddenPhrase(randPhrase, guessed):
    p = ''
    for g in randPhrase:
        if (g in letters) and (g not in guessed):
            p = p+' _ '
        else:
            p = p+g
    return p

def wheelSpin():
    randSpin = random.choice(wheel_segments)
    return randSpin   

print('WHEEL OF FORTUNE')

player_count = 3

player_names= [Player(input('Enter the name for player #{}'.format(i+1))) for i in range(player_count)]

players = player_names

playerIndex = 0
                        
play = True
pre_rounds_left = 2
final_round = 1


while(play):
    
    while(pre_rounds_left > 0):

        randPhrase = getphrase()

        guessed = []

        def requestPlayerGuess(player, guessed):
            while True: 
                guess = player.getGuess(hiddenPhrase(randPhrase, guessed), guessed)
                guess = guess.upper() 
                if len(guess) == 1: 
                    if guess not in letters: 
                        print('Guess a letter.')
                        continue
                    elif guess in guessed:
                        print('{} has already been guessed'.format(guess))
                        continue
                    elif guess in vowels and player.prizeMoney < vowel_cost: 
                            print('Need $250 to buy a vowel.')
                            continue
                    else:
                        return guess
                else:
                    return guess

        while True:
            player = players[playerIndex]
            prize = wheelSpin()
            print('{} spins {}'.format(player.name, prize))
            if prize == 'Bankrupt':
                player.goBankrupt()
            elif prize == 'Lose a Turn':
                pass 
            elif prize >= 50 or prize <= 950:
                guess = requestPlayerGuess(player, guessed)
                if len(guess) == 1:
                    guessed.append(guess)

                    print('{} guesses "{}"'.format(player.name, guess))

                    if guess in vowels:
                        player.prizeMoney -= vowel_cost

                    count = randPhrase.count(guess) 
                    if count > 0:
                        if count == 1:
                            print("There is one {}".format(guess))
                        else:
                            print("There are {} {}'s".format(count, guess))

                        player.addMoney(prize)

                        if hiddenPhrase(randPhrase, guessed) == randPhrase:
                            roundWinner = player
                            pre_rounds_left -= 1
                            print('{} won the round! The phrase was {}'.format(roundWinner.name, randPhrase))  
                            break
                        continue 

                    elif count == 0:
                        print("There is no {}".format(guess))
                else: 
                    if guess == randPhrase: 
                        roundWinner = player
                        pre_rounds_left -= 1
                        player.addMoney(prize)
                        print('{} won the round! The phrase was {}'.format(roundWinner.name, randPhrase))  
                        break                                              
                    else:
                        print('{} was not the phrase'.format(guess))

            playerIndex = (playerIndex + 1) % len(players)
            
    while (pre_rounds_left == 0 and final_round == 1):
        
        print('Welcome to the final round!')
        
        print('In this final round, you can select only 3 more consonants and 1 more vowel.')
        
        print('When you are out of letters to guess, you can try to guess the final phrase on last time')
        
        print('If you are able to guess the final phrase then you will win the GRAND PRIZE of $100,000!')
        
        
        randPhrase = getphrase()

        guessed = ['R', 'S', 'T', 'L', 'N', 'E']
        
        consonants_left = 3
        
        vowels_left = 1
        
        
        def requestPlayerGuess(player, guessed):
            while True: 
                guess = player.getGuess(hiddenPhrase(randPhrase, guessed), guessed)
                guess = guess.upper() 
                if len(guess) == 1: 
                    if guess not in letters: 
                        print('Guess a letter.')
                        continue
                    elif guess in guessed:
                        print('{} has already been guessed'.format(guess))
                        continue
                    elif guess in vowels and vowels_left == 0: 
                        print('You have no more vowels')
                        continue
                    elif consonants_left == 0:
                        print('You have no more consonants left')
                        continue
                    else:
                        return guess
                        
                else:
                    return guess

        while True:
            player = players[playerIndex]
            prize = 100000
           
            guess = requestPlayerGuess(player, guessed)
            if len(guess) == 1:
                guessed.append(guess)
                consonants_left -= 1

                print('{} guesses "{}"'.format(player.name, guess))

                if guess in vowels:
                    player.prizeMoney -= vowel_cost
                    vowels_left -= 1

                count = randPhrase.count(guess) 
                if count > 0:
                    if count == 1:
                        print("There is one {}".format(guess))
                    else:
                        print("There are {} {}'s".format(count, guess))

                    if hiddenPhrase(randPhrase, guessed) == randPhrase:
                        gameWinner = player
                        player.addMoney(prize)
                        print('{} won the Game! The phrase was {}'.format(roundWinner.name, randPhrase))  
                        break
                    continue 

                elif count == 0:
                    print("There is no {}".format(guess))
            else: 
                if guess == randPhrase: 
                    gameWinner = player
                    player.addMoney(prize)
                    print('You won the Game! The phrase was {}'.format(roundWinner.name, randPhrase))
                    final_round -= 1
                    break                                              
                else:
                    print('{} was not the phrase'.format(guess))
                    print('Sorry, you lost the game')
                    final_round -= 1
                    break

```
